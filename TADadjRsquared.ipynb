```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas
from sklearn import metrics
import urllib2
```

```python
t = np.loadtxt(urllib2.urlopen('http://bx.psu.edu/~lua137/OnTAD/rebuttal/testadjr2alltest_KRnorm_new_TopDom.txt'))
t=t[:,0:150]
```

```python
plt.figure(5)
fig,ax = plt.subplots(1)
ax.plot(t[0,:],c='#66c2a5',label='OnTAD',linewidth=2)
ax.plot(t[2,:],c='#a6d854',label='DomainCaller',linewidth=2)
ax.plot(t[1,:],c='#fc8d62',label='Arrowhead',linewidth=2)
ax.plot(t[3,:],c='#8da0cb',label='rGMAP',linewidth=2)
ax.plot(t[4,:],c='#e78ac3',label='TADtree',linewidth=2)
ax.plot(t[5,:],c='#ffd92f',label='TopDom',linewidth=2)
ax.legend(loc="upper right",prop={'size': 10})
#ax.axes.get_xaxis().set_visible(False)
plt.ylabel('TAD$adjR^2$', {'color': 'k', 'fontsize': 20})
plt.xlabel('Genomic Distance', {'color': 'k', 'fontsize': 20})
plt.yticks(color='k',size=14)
plt.xticks((0,50,100,150),('0','500Kb','1Mb','1.5Mb'),color='k',size=15)

plt.savefig('/Users/linan/Desktop/testadjr2all_TopDom.png',dpi=300, transparent=True, bbox_inches='tight')
plt.show()
#plt.close()
```
